```python
# Ejercicio #5.
# Manipulación de Texto:
# Crea una función que tome una columna de descripciones de productos y devuelva una nueva columna con las palabras
# más comunes de cada descripción.

import pandas as pd
from collections import Counter

data = {
    'Descripcion': [
        'Este es un producto increible y util',
        'producto de alta calidad',
        'Gran oferta en este producto'
    ]
}

df = pd.DataFrame(data)

# Definir la función para obtener las palabras más comunes
def palabras_comunes(descripcion, num_palabras=3):
    palabras = descripcion.lower().split()
    contador = Counter(palabras)
    palabras_comunes = contador.most_common(num_palabras)
    return ', '.join([palabra for palabra, _ in palabras_comunes])

# Aplicar la función para crear una nueva columna
df['PalabrasComunes'] = df['Descripcion'].apply(palabras_comunes)

print(df)
```

```python
# Ejercicio #2.
# Agregación de Datos:
# A partir del DataFrame de ventas, calcula la media de ventas, la suma de unidades vendidas y la cantidad 
# de productos únicos vendidos por categoría.

import pandas as pd

data = pd.read_csv("C:/Users/Administrador/Desktop/Python/Analisis_de_datos/Archivos/ventas.csv")
 
df = pd.DataFrame(data)

# Calcula la media de ventas, la suma de unidades vendidas y la cantidad de productos unicos vendidos por categoria
resultados = df.groupby('categoria').agg({'ventas': 'mean', 'cantidad': 'sum', 'producto': 'nunique'})

print(resultados)
```

```python
# Ejercicio #3.
# Agregación con Funciones Personalizadas:
# Define una función personalizada que calcule la diferencia entre la venta máxima y la venta mínima para
# cada categoría, y aplícala utilizando groupby().

import pandas as pd

data = pd.read_csv("C:/Users/Administrador/Desktop/Python/Analisis_de_datos/Archivos/ventas.csv")
 
df = pd.DataFrame(data)

# Define la función personalizada para calcular la diferencia entre la venta máxima y la venta mínima
def diferencia_ventas(series):
    return series.max() - series.min()

# Aplica la función personalizada utilizando groupby() en la columna 'categoria'
resultados = df.groupby('categoria').agg({'ventas': diferencia_ventas})

print(resultados)
```

```python
# Ejercicio #4.
# Agrupación por Múltiples Columnas:
# Agrupa los datos de ventas por año y mes, y calcula la suma de ventas para cada combinación de año y mes.

import pandas as pd
import datetime

data = pd.read_csv("C:/Users/Administrador/Desktop/Python/Analisis_de_datos/Archivos/Ventas.csv")
 
df = pd.DataFrame(data)

# Convierte la columna 'fecha_venta' a tipo datetime.
df['fecha_venta'] = pd.to_datetime(df['fecha_venta'])

# Agrupa los datos por año y mes, y calcula la suma de ventas para cada combinación de año y mes
resultados = df.groupby([df['fecha_venta'].dt.year, df['fecha_venta'].dt.month]).agg({'ventas': 'sum'})

print(resultados)
```

```python
# Ejercicio #3.
# Segmentación y Filtrado por Grupo:
# Itera sobre un DataFrame de estudiantes agrupados por género y selecciona los estudiantes con las calificaciones
# más altas en cada grupo. 

import pandas as pd

# DataFrame 'estudiantes' con información sobre los estudiantes
data = {
    'Nombre': ['Estudiante A', 'Estudiante B', 'Estudiante C', 'Estudiante D', 'Estudiante E', 'Estudiante F'],
    'Genero': ['M', 'F', 'M', 'F', 'M', 'F'],
    'Calificaciones': [85, 92, 78, 95, 88, 90]
}

estudiantes = pd.DataFrame(data)

# Agrupamos los estudiantes por genero.
grupos = estudiantes.groupby('Genero')

# Iteramos sobre cada grupo y seleccionamos los estudiantes con las calificaciones mas altas.
mejores_calificaciones = []

for genero, grupo in grupos:
    estudiante_mejor_calificacion = grupo.loc[grupo['Calificaciones'].idxmax()]
    mejores_calificaciones.append(estudiante_mejor_calificacion)

resultados = pd.DataFrame(mejores_calificaciones)

print(resultados)
```

```python
# Ejercicio #4.
# Análisis Temporal por Grupo:
# Aplica la iteración para calcular la tendencia de ventas a lo largo del tiempo en diferentes categorías de productos.

import pandas as pd

data = pd.read_csv("C:/Users/Administrador/Desktop/Python/Analisis_de_datos/Archivos/ventas.csv")

ventas = pd.DataFrame(data)

# Convertir la columna de fecha_ventas a tipo datetime.
ventas['fecha_venta_venta'] = pd.to_datetime(ventas['fecha_venta'])  

# Agrupamos las ventas por producto.
grupos = ventas.groupby('producto')

# Creamos un DataFrame vacio para almacenar los resultados.
resultados = pd.DataFrame(columns=['producto', 'fecha_venta', 'tendencia'])

# Iteramos sobre cada grupo y calculamos la tendencia de ventas a lo largo del tiempo.
for producto, grupo in grupos:
    grupo = grupo.sort_values(by='fecha_venta')  # Ordenar por fecha_venta.
    grupo['tendencia'] = grupo['ventas'].diff()  # Calcular la diferencia de ventas respecto al día anterior.
    grupo = grupo.dropna()  # Eliminar la primera.
    resultados = resultados.append(grupo)

print(resultados)
```

```python
# Ejercicio #5.
# Personalización de Análisis por Grupo:
# Utiliza la iteración para realizar un análisis personalizado sobre un conjunto de datos, como identificar
# productos con ventas estacionales y sugerir estrategias de marketing.

import pandas as pd

data = pd.read_csv("C:/Users/Administrador/Desktop/Python/Analisis_de_datos/Archivos/ventas.csv")

ventas = pd.DataFrame(data)
ventas['fecha_venta'] = pd.to_datetime(ventas['fecha_venta'])  # Convertir la columna de fecha_ventas a tipo datetime

# Definimos una función para identificar productos con ventas estacionales
def analizar_ventas_estacionales(grupo):
    variacion_ventas = grupo['ventas'].max() - grupo['ventas'].min()
    
    if variacion_ventas > 500:
        return "Ventas estacionales"
    else:
        return "Ventas consistentes"

# Agrupamos las ventas por producto.
grupos = ventas.groupby('producto')

# Iteramos sobre cada grupo y analizamos las ventas.
for producto, grupo in grupos:
    analisis = analizar_ventas_estacionales(grupo)
    print(f"Producto {producto}: {analisis}")
    
    # Sugerir estrategias de marketing basadas en el analisis.
    if analisis == "Ventas estacionales":
        print("Sugerencia de marketing: Aprovechar los picos de ventas con campañas promocionales durante los periodos de alta demanda.")
    else:
        print("Sugerencia de marketing: Mantener una presencia constante en el mercado y enfocarse en la felicidad de los clientes.")
    
    print("\n")
```
